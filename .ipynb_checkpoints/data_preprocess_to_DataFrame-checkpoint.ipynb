## Data Process

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import numpy as np
import os
```

```python
sample_image = load_img('data/chest_xray/train/NORMAL/IM-0115-0001.jpeg')
display(plt.imshow(sample_image))
```

```python
sample_image_array = img_to_array(sample_image)
```

```python
sample_image_array.shape
```

```python
train_data_dir = 'data/chest_xray/train'
val_data_dir = 'data/chest_xray/val'
test_data_dir = 'data/chest_xray/test'

# Get all the data in the directory data/train (5216 images), and reshape them
train_generator = ImageDataGenerator().flow_from_directory(
        train_data_dir, 
        target_size=(64, 64), 
        batch_size=5216
        )

# Get all the data in the directory data/val (16 images), and reshape them
val_generator = ImageDataGenerator().flow_from_directory(
        val_data_dir, 
        target_size=(64, 64), 
        batch_size=16
        )

# Get all the data in the directory data/test (624 images), and reshape them
test_generator = ImageDataGenerator().flow_from_directory(
        test_data_dir, 
        target_size=(64, 64), 
        batch_size=624
        )

# Create the datasets
train_images, train_labels = next(train_generator)
val_images, val_labels = next(val_generator)
test_images, test_labels = next(test_generator)
```

```python
val_labels2 = val_labels[:,1]
val_labels2.shape
```

```python
val_labels2[5]
```

```python
val_images2 = val_images.reshape(16, -1).astype('float32')
val_images2.shape
```

```python
val_images2 /= 255.
```

```python
df_val_labels2 = pd.DataFrame(val_labels2, columns=['pneumonia'])
df_val_images2 = pd.DataFrame(val_images2)
df_val = pd.concat([df_val_labels2, df_val_images2], axis=1)
print(df_val.info())
df_val.head()
```

```python
test_labels2 = test_labels[:,1]
test_images2 = test_images.reshape(624, -1).astype('float32')
test_images2 /= 255.
df_test_labels2 = pd.DataFrame(test_labels2, columns=['pneumonia'])
df_test_images2 = pd.DataFrame(test_images2)
df_test = pd.concat([df_test_labels2, df_test_images2], axis=1)
print(df_test.info())
df_test.head()
```

```python
train_labels2 = train_labels[:,1]
train_images2 = train_images.reshape(5216, -1).astype('float32')
train_images2 /= 255.
df_train_labels2 = pd.DataFrame(train_labels2, columns=['pneumonia'])
df_train_images2 = pd.DataFrame(train_images2)
df_train = pd.concat([df_train_labels2, df_train_images2], axis=1)
print(df_train.info())
df_train.head()
```

```python
df_all = df_train.append(df_val, ignore_index=True)
df_all = df_all.append(df_test, ignore_index=True)
print(df_all.info())
df_all.head(20)
```
